```python
import numpy as np
from scipy import signal
from scipy.fft import fft, fftshift
import matplotlib.pyplot as plt

window = signal.windows.flattop(51)
window_2 = signal.windows.blackmanharris(51)
window_3 = signal.windows.triang(51)
window_4 = signal.windows.boxcar(51)
window_5 = signal.windows.hamming(51)

plt.figure(figsize=(20, 10))
A = fft(window, 2048) / (len(window)/2.0)
A_2 = fft(window_2, 2048) / (len(window_2)/2.0)
A_3 = fft(window_3, 2048) / (len(window_3)/2.0)
A_4 = fft(window_4, 2048) / (len(window_4)/2.0)
A_5 = fft(window_5, 2048) / (len(window_5)/2.0)
freq = np.linspace(-0.5, 0.5, len(A))
response = 20 * np.log10(np.abs(fftshift(A / abs(A).max())))
response_2 = 20 * np.log10(np.abs(fftshift(A_2 / abs(A_2).max())))
response_3 = 20 * np.log10(np.abs(fftshift(A_3 / abs(A_3).max())))
response_4 = 20 * np.log10(np.abs(fftshift(A_4 / abs(A_4).max())))
response_5 = 20 * np.log10(np.abs(fftshift(A_5 / abs(A_5).max())))
plt.plot(freq, response,label='Flattop window')
plt.plot(freq, response_2,label='Blackman-Harris window')
plt.plot(freq, response_3,label='Triangular window')
plt.plot(freq, response_4,label='Boxcar window')
plt.plot(freq, response_5,label='Hamming window')
plt.axis([-0.5, 0.5, -120, 0])
plt.title("Frequency response of the Flattop window, Blackman-Harris window, Triangular window, and Hamming window")
plt.ylabel("Normalized magnitude [dB]")
plt.xlabel("Normalized frequency [cycles per sample]")
plt.grid()
plt.legend()
plt.show()

```
