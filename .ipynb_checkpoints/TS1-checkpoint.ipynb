```python
#Trabajo semanal 1
Para este trabajo se preparo un programa para diseñar y parametrizar señales. En este caso se busco graficar una funcion senoidal
Tomando de referencia la siguiente funcion:
/begin{equation}
    X(t)= A \sen(\omega t + \phi)
/end{equation}

Seguido de fijar una N cantidad de muestras digitalizadas y una Frecuencia de muestro para el ADC.
Finalmente Graficar dicha funcion.

Como agregado para el experimento, se probó con distintas frecuencias
```

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Mar  7 15:30:48 2025

@author: Joaquin

------------------
Consigna:
    programar una función que genere señales senoidales y que permita parametrizar:
        la amplitud máxima de la senoidal (volts)
        su valor medio (volts)
        la frecuencia (Hz)
        la fase (radianes)
        la cantidad de muestras digitalizada por el ADC (# muestras)
        la frecuencia de muestreo del ADC.
        
        es decir que la función que uds armen debería admitir se llamada de la siguiente manera

    tt, xx = mi_funcion_sen( vmax = 1, dc = 0, ff = 1, ph=0, nn = N, fs = fs)
    Recuerden que tanto xx como tt deben ser vectores de Nx1
"""


import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl


def funcion_sen(Vmax,dc,ff,ph, nn,fs):
    Ts=1/fs
    t=np.linspace(0,(nn-1)*Ts,nn).flatten()
    argg=2*np.pi*ff*t
    return [t,Vmax * np.sin(argg) + ph]
    
N=1000 #La cantidad de muestras digitalizada por el ADC
Fs=1000 #Frecuencia de muestreo del ADC
F0=10
                                 
tt, xx = funcion_sen(1,0,F0,0,N,Fs)


plt.plot(tt,xx)
plt.xlabel("tiempo (t)")
plt.ylabel("Volts")
plt.title("Grafico de mi funcion senoidal")
plt.axhline(0,color="black",linewidth=1)
plt.axvline(0,color="black",linewidth=1)
plt.grid(True,linestyle=":",alpha=0.6)
plt.show()
```

```python
Ademas de probar con diferentes frecuencias, se me ocurrio intentar graficar una señal cuadrada utilizando la serie de Fourier de la siguiente funcion:
\begin{equation}
\[
f(x) = 
\begin{cases} 
-1 & -\pi < x < 0 \\ 
1 & 0 < x < \pi 
\end{cases}
\]
\end{equation}

Llegando asi a la Serie:
\[
\frac{4}{\pi} \sum_{n=1}^{N-1} \frac{\sin\left((2n - 1) \cdot 2\pi \cdot f \cdot t\right)}{2n - 1}
\]

obteniendo:
```

```python
import numpy as np
import matplotlib.pyplot as plt


def Funcion_cuadrada(Vmax,ff,ph,nn,fs):
    Ts = 1/fs
    t=np.linspace(0,(nn-1)*Ts,nn).flatten()
    N=10000
    return[t,Vmax*((4/np.pi)*sum(np.sin((2*n-1)*2*np.pi*ff*t)/(2*n-1) for n in range(1,N-1)))]

N=1000 #La cantidad de muestras digitalizada por el ADC
Fs=1000 #Frecuencia de muestreo del ADC
F0=10
                                 
tt_1,xx_1 = Funcion_cuadrada(2,F0,0,N, Fs)

plt.plot(tt_1,xx_1)
plt.xlabel("tiempo (t)")
plt.ylabel("Volts")
plt.title("Grafico de mi funcion senoidal")
plt.axhline(0,color="black",linewidth=1)
plt.axvline(0,color="black",linewidth=1)
plt.grid(True,linestyle=":",alpha=0.6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

def Funcion_cuadrda(Vmax,ff,ph, nn,fs):
    Ts=1/fs
    t=linspace(0,(nn-1)*Ts,nn).flatten()
    return[t,signal.square(2 * np.pi * ff * t + ph)]



N=1000 #La cantidad de muestras digitalizada por el ADC
Fs=1000 #Frecuencia de muestreo del ADC
F0=10
                                 
tt_1,xx_1 = Funcion_cuadrada(2,F0,0,N, Fs)

plt.plot(tt_1,xx_1)
plt.xlabel("tiempo (t)")
plt.ylabel("Volts")
plt.title("Grafico de mi funcion cuadrada")
plt.axhline(0,color="black",linewidth=1)
plt.axvline(0,color="black",linewidth=1)
plt.grid(True,linestyle=":",alpha=0.6)
plt.show()
```
